```python
csv_file1 = "../project/carlos_cleaned.csv"
country_ranking1 = pd.read_csv(csv_file1)
country_ranking2 = pd.DataFrame(country_ranking1)
country_rankings = country_ranking2.drop(columns = ['Unnamed: 0'], axis = 1)
country_rankings.head()
```

```python
csv_file2 = "../project/final_db_goverment.csv"
government_df = pd.read_csv(csv_file2)
government_dfs = pd.DataFrame(government_df)
new_government_data = government_dfs.drop(columns = ['Unnamed: 0'], axis = 1)
no_na_government_data = new_government_data.dropna(how = 'any')
no_na_government_data.head()
```

Use pandas to load csv converted DataFrame into database

```python
rds_connection_string = "postgres:postgres@localhost:5432/Economic-_datasets"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names() # check to see table name, make sure they are consistent with SQL
```

```python
no_na_government_data.to_sql(name='no_na_government_data', con=engine, if_exists='append', index=False) #append the data into the SQL
```

```python
country_rankings.to_sql(name='country_rankings', con=engine, if_exists='append', index=False)#append the data into the SQL
```

```python
country_profile.to_sql(name='country_profile', con=engine, if_exists='append', index=False)#append the data into the SQL
```
